##### Implement the Dot product of two vectors

```python
def dot_product(x, y):
    return sum([a * b for a, b in zip(x, y)])
```

```python
dot_product([5, 0, -3], [2, -5, 6])
```

##### Implement the Hadamard product of two vectors

```python
def hadamard_product(x, y):
    return [a * b for a, b in zip(x, y)]
```

```python
hadamard_product([5, 0, -3], [2, -5, 6])
```

_Outliers (IQR)_

Let's also find out outliers in the dataset by using IQR.

##### Weight

```python
q1, q3 = np.percentile(sorted(gender_data['Weight']), [25, 75])
iqr = q3 - q1

lower_bound = q1 - (1.5 * iqr)
upper_bound = q3 + (1.5 * iqr)

(lower_bound, upper_bound)
```

This shows that any weight below $55$ or above $77.7$ can be considered to be outliers. These outlier observations (eight in total) are displayed below.

```python
gender_data[(gender_data['Weight'] < lower_bound) | (gender_data['Weight'] > upper_bound)]
```

##### Height

```python
q1, q3 = np.percentile(sorted(gender_data['Height']), [25, 75])
iqr = q3 - q1

lower_bound = q1 - (1.5 * iqr)
upper_bound = q3 + (1.5 * iqr)

(lower_bound, upper_bound)
```

This shows that any height below $58.8$ or above $264.2$ can be considered to be outliers. The single outlier observation is displayed below.

```python
gender_data[(gender_data['Height'] < lower_bound) | (gender_data['Height'] > upper_bound)]
```

Please note that this observation was considered to be outlier in terms of weight too.

Overall, this illustrates that weight data has more outliers by IQR than height data does. On the other hand, the number of outliers is relatively low, considering that the dataset contains $10000$ observations. 

_Outliers (Z-score)_

Now let's find out outliers by using Z-score approach. In this case, we mark an observation as an outlier if the z-score for that data point is outside of $3$ standard deviations.

##### Weight

```python
from scipy import stats

gender_data[np.abs(stats.zscore(gender_data['Weight'])) > 3]
```

This demonstrates that seven data points are considered to be outlier in terms of weight. Interestingly, this method didn't classify the observation with $7294$ as outlier.

##### Height

```python
gender_data[np.abs(stats.zscore(gender_data['Height'])) > 3]
```

We can see that in this case we have two outlier observations in terms of height. The observation with index $9285$ was indicated to be outlier by IQR method.

We can conclude that overall there are two outlier data points that have anomaly values for both height and weight. These observations are a male with weight nearly $79$ and height $270$, and a female with weight about $54.3$ and height $64.7$.

Seemingly, there are a few outlier observations in the dataset. There is a significant difference between male and female data of weights and heights. Usually the results are considerably higher for males than for females. Also, there is a strond linear relationship between weight and height.
